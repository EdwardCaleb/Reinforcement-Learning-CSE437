# Gaussian Process

```python
import torch
import gpytorch
import numpy as np

# ============================================================
# ---- Base Exact GP model -------------------
# ============================================================

class ExactGPModel(gpytorch.models.ExactGP):
    def __init__(self, train_x, train_y, likelihood, kernel='RBF', ard_dims=None):
        super().__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ConstantMean()

        # Choose kernel type
        if kernel == 'RBF':
            base_kernel = gpytorch.kernels.RBFKernel(ard_num_dims=ard_dims)
        elif kernel == 'Matern':
            base_kernel = gpytorch.kernels.MaternKernel(nu=2.5, ard_num_dims=ard_dims)
        elif kernel == 'RQ':
            base_kernel = gpytorch.kernels.RationalQuadraticKernel(ard_num_dims=ard_dims)
        else:
            raise ValueError(f"Unsupported kernel type: {kernel}")

        self.covar_module = gpytorch.kernels.ScaleKernel(base_kernel)

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)


# ============================================================
# ---- Individual GP Manager: handles data + normalization ----
# ============================================================

class GPManager:
    def __init__(self, kernel='RBF', lr=0.03, iters=1000):
        self.kernel = kernel
        self.lr = lr
        self.iters = iters
        self.trained = False
        self.X_train = []
        self.Y_train = []

        self.likelihood = None
        self.model = None


    # ----------------------------- #
    #        FIT / INITIAL TRAIN    #
    # ----------------------------- #
    def fit(self, X, Y):
        X = torch.tensor(X, dtype=torch.float32)
        Y = torch.tensor(Y, dtype=torch.float32)

        self.X_train = X.clone()
        self.Y_train = Y.clone()

        self.retrain()

    def retrain(self):
        # Normalize the input/output and then train the model
        self._compute_normalization()
        self._train_model()

    # ----------------------------- #
    #       ADD NEW DATA POINTS     #
    # ----------------------------- #
    def add_data(self, X_new, Y_new, retrain=True):
        """
        Add new data (single or batch) to GP training set.
        Optionally retrain the model.
        """
        X_new = torch.tensor(X_new, dtype=torch.float32)
        Y_new = torch.tensor(Y_new, dtype=torch.float32).flatten()  # ensure 1D

        # Also flatten stored Y to ensure shape consistency
        if self.Y_train.ndim > 1:
            self.Y_train = self.Y_train.flatten()

        # Append new data
        self.X_train = torch.cat([self.X_train, X_new], dim=0)
        self.Y_train = torch.cat([self.Y_train, Y_new], dim=0)

        if retrain:
            self.retrain()

    # ----------------------------- #
    #         INTERNAL UTILS        #
    # ----------------------------- #
    def _compute_normalization(self):
        self.X_mean, self.X_std = self.X_train.mean(0), self.X_train.std(0)
        self.Y_mean, self.Y_std = float(self.Y_train.mean(0)), float(self.Y_train.std(0))

        Xn = (self.X_train - self.X_mean) / self.X_std
        Yn = (self.Y_train - self.Y_mean) / self.Y_std

        self.Xn, self.Yn = Xn, Yn

    def dataset(self):
        # Extract training data (already stored inside the wrapper)
        X_train = self.X_train.numpy()
        Y_train = self.Y_train.numpy()
        return X_train, Y_train


    def _train_model(self):
        self.likelihood = gpytorch.likelihoods.GaussianLikelihood()
        self.likelihood.noise_covar.initialize(noise=1e-3)
        self.model = ExactGPModel(self.Xn, self.Yn, self.likelihood,
                                  kernel=self.kernel, ard_dims=self.X_train.shape[-1])

        # -------------------------------------------------
        # Train the model
        # -------------------------------------------------
        self.train_gp(self.model, self.likelihood, self.Xn, self.Yn)
        self.trained = True

    def train_gp(self, model, likelihood, x, y):
        model.train(); likelihood.train()
        opt = torch.optim.Adam(model.parameters(), lr=self.lr)
        mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)
        for _ in range(self.iters):
            opt.zero_grad()
            out = model(x)
            loss = -mll(out, y)
            loss.backward()
            opt.step()
        model.eval(); likelihood.eval()

    def predict(self, X):
        if not self.trained:
            raise RuntimeError("GP has not been trained yet.")

        # Normalize the input
        X = torch.as_tensor(X, dtype=torch.float32)
        Xn = (X - self.X_mean) / self.X_std
        with torch.no_grad(), gpytorch.settings.fast_pred_var():
            pred = self.likelihood(self.model(Xn))

            # Unonormalize the output
            mean = pred.mean.cpu().numpy() * self.Y_std + self.Y_mean
            var = pred.variance.cpu().numpy() * (self.Y_std ** 2)
        # Return the unnormalized output
        return mean, var
```

# Setup environment and collect initial data

```python
import gymnasium as gym
# import gym
import numpy as np
import torch
import matplotlib.pyplot as plt

# --------------------------------------------------------------
# 1.  Environment setup
# --------------------------------------------------------------
env = gym.make("MountainCarContinuous-v0", max_episode_steps=500)
n_state = env.observation_space.shape[0]    # [position, velocity]
n_action = env.action_space.shape[0]        # 1 continuous action
n_output = 2                                # Δpos, Δvel


# --------------------------------------------------------------
# 2.  Collect initial random data
# --------------------------------------------------------------
data = {'X': [], 'Y': []}

obs, _ = env.reset(seed=0)
for _ in range(200):  # collect 500 transitions
    a = env.action_space.sample()  # random continuous action in [-1,1]
    x_next, r, term, trunc, _ = env.step(a)
    data['X'].append(np.concatenate([obs, a]))   # [pos, vel, action]
    data['Y'].append(x_next - obs)               # Δstate
    obs = x_next
    if term or trunc:
        obs, _ = env.reset()

X = np.array(data['X'])[::3]
Y = np.array(data['Y'])[::3]
print(f"Collected {len(X)} transitions.")


# --------------------------------------------------------------
# 3.  Train GP dynamics models
# --------------------------------------------------------------
gps = [GPManager(kernel='RBF', iters=300) for _ in range(n_output)]

for d in range(n_output):
    gps[d].fit(X, Y[:, d])
    print(f"Trained GP for Δstate[{d}] with {len(X)} samples.")
```

```python
def gp_dynamics_step(state, action, gps):
    """
    Propaga la dinámica 1 paso hacia adelante usando los GPs.
    state: np.array([pos, vel])
    action: float o array escalar
    gps: lista [gp_Δpos, gp_Δvel]
    """
    if isinstance(action, np.ndarray):
        a = float(action.squeeze())
    else:
        a = float(action)

    x_in = np.concatenate([state, [a]])[None, :]   # shape (1, 3)

    dpos_mean, _ = gps[0].predict(x_in)
    dvel_mean, _ = gps[1].predict(x_in)

    next_state = state + np.array([dpos_mean[0], dvel_mean[0]])
    return next_state
```

```python
def mpc_action(state,
               gps,
               horizon=30,
               n_samples=500,
               action_low=-1.0,
               action_high=1.0,
               lambda_action=0.001,
               goal_pos=0.45):
    """
    MPPI-Simplified (Random Shooting) con dinámica dada por los GPs.
    - state: np.array([pos, vel]) estado actual
    - gps: lista [GP_Δpos, GP_Δvel]
    - horizon: longitud del horizonte de planificación
    - n_samples: número de trayectorias aleatorias
    - action_low, action_high: límites de acción (MountainCarContinuous)
    - lambda_action: peso de regularización de la acción
    - goal_pos: posición objetivo (≈ 0.45 para MountainCarContinuous)

    Devuelve:
    - best_action: np.array([a]) para pasar directo a env.step()
    - best_seq: secuencia de acciones del mejor roll-out
    - best_cost: coste asociado
    """
    # Muestreamos secuencias aleatorias de acciones
    action_seqs = np.random.uniform(
        low=action_low,
        high=action_high,
        size=(n_samples, horizon)
    )

    costs = np.zeros(n_samples)

    for k in range(n_samples):
        s = state.copy()
        cost = 0.0

        for t in range(horizon):
            a = action_seqs[k, t]

            # Propagar con el modelo GP
            s = gp_dynamics_step(s, a, gps)
            pos = s[0]
            vel = s[1]

            # Coste de etapa: estar lejos de la cima + penalización de esfuerzo
            cost += 0.2*(goal_pos - pos) ** 2 - 500*(vel**2) + 0.01*lambda_action * (a ** 2)

        costs[k] = cost

    # Elegimos la trayectoria con menor coste
    best_idx = np.argmin(costs)
    best_seq = action_seqs[best_idx]
    best_action = best_seq[0]

    return np.array([best_action], dtype=np.float32), best_seq, costs[best_idx]
```

    - In Model Based Reinforcement Learning, the algorithm first learn the dynamic system and uses it afterwards to predict the best way to reach some goal, doesn't matter if the goal changes, as long as the dynamical behavior doesn't change, it is still possible to reuse the model to achieve a new goal, making it much more data-efficient than Q-Leaning algorithm.

* Describe the downside of MBRL over Q-learning

The other big downside is that MBRL is too costly computationally talking, needs to run hundreds of trajectories, evaluate them and optimize them to only take the first action, something that in Q-Learning isn't necessary since it learns the Q values first and then only consults the table to know which decision to take.

```python
def gp_dynamics_step_mean(state, action, gps):
    """
    Un paso de dinámica usando SOLO la media del GP (modelo determinista).
    state: np.array([pos, vel])
    action: escalar o array de tamaño 1
    """
    if isinstance(action, np.ndarray):
        a = float(action.squeeze())
    else:
        a = float(action)

    x_in = np.concatenate([state, [a]])[None, :]   # shape (1, 3)

    dpos_mean, _ = gps[0].predict(x_in)
    dvel_mean, _ = gps[1].predict(x_in)

    next_state = state + np.array([dpos_mean[0], dvel_mean[0]])
    return next_state


def gp_dynamics_step_mean_var(state, action, gps):
    """
    Un paso de dinámica usando media y varianza del GP.
    Devuelve:
      - next_state_mean: np.array([pos, vel])
      - next_state_var:  np.array([var_Δpos, var_Δvel])
    """
    if isinstance(action, np.ndarray):
        a = float(action.squeeze())
    else:
        a = float(action)

    x_in = np.concatenate([state, [a]])[None, :]

    dpos_mean, dpos_var = gps[0].predict(x_in)
    dvel_mean, dvel_var = gps[1].predict(x_in)

    next_state_mean = state + np.array([dpos_mean[0], dvel_mean[0]])
    next_state_var = np.array([dpos_var[0], dvel_var[0]])  # varianza en Δestado

    return next_state_mean, next_state_var
```

```python
def mpc_action(
    state,
    gps,
    horizon=30,
    n_samples=500,
    action_low=-1.0,
    action_high=1.0,
    lambda_action=1.0,
    goal_pos=0.45,
    use_variance=False,
    beta_uncertainty=1.0
):
    """
    MPPI-Simplified (Random Shooting) con opción de incluir varianza del GP.

    state: np.array([pos, vel])
    gps: lista [GP_Δpos, GP_Δvel]
    use_variance: si True, se penaliza la varianza del modelo a lo largo de la trayectoria
    beta_uncertainty: peso de la penalización de incertidumbre
    """
    # Secuencias de acciones aleatorias
    action_seqs = np.random.uniform(
        low=action_low,
        high=action_high,
        size=(n_samples, horizon)
    )

    costs = np.zeros(n_samples)

    for k in range(n_samples):
        s = state.copy()
        cost = 0.0

        for t in range(horizon):
            a = action_seqs[k, t]

            if use_variance:
                s_next, var_next = gp_dynamics_step_mean_var(s, a, gps)
            else:
                s_next = gp_dynamics_step_mean(s, a, gps)
                var_next = np.array([0.0, 0.0])  # sin penalización

            pos, vel = s_next
            s = s_next

            # --- COSTE BASE (el que comentaste que te funcionó bien) ---
            stage_cost = (
                0.2 * (goal_pos - pos) ** 2
                - 500.0 * (vel ** 2)
                + 0.01 * lambda_action * (a ** 2)
            )

            # --- PENALIZACIÓN POR INCERTIDUMBRE DEL GP ---
            # Usamos la suma de varianzas de Δpos y Δvel como medida de incertidumbre:
            uncertainty = np.sum(var_next)  # escala positiva

            if use_variance:
                stage_cost += beta_uncertainty * uncertainty

            cost += stage_cost

        costs[k] = cost

    # Elegimos la mejor trayectoria (menor coste)
    best_idx = np.argmin(costs)
    best_seq = action_seqs[best_idx]
    best_action = best_seq[0]

    return np.array([best_action], dtype=np.float32), best_seq, costs[best_idx]
```

```python
def run_episode(env, gps, use_variance=False, beta_uncertainty=1.0,
                horizon=30, n_samples=500, goal_pos=0.45, seed=None):
    """
    Ejecuta un episodio completo usando MPC con/ sin incertidumbre.
    Devuelve:
      positions: lista de posiciones en el tiempo
      rewards:   lista de recompensas reales del entorno
      steps:     número de pasos hasta terminar el episodio
    """
    if seed is not None:
        obs, _ = env.reset(seed=seed)
    else:
        obs, _ = env.reset()

    positions = [obs[0]]
    rewards = []
    done = False

    for t in tqdm(range(env._max_episode_steps)):  # o 500
        a, _, _ = mpc_action(
            state=obs,
            gps=gps,
            horizon=horizon,
            n_samples=n_samples,
            action_low=env.action_space.low[0],
            action_high=env.action_space.high[0],
            lambda_action=1.0,
            goal_pos=goal_pos,
            use_variance=use_variance,
            beta_uncertainty=beta_uncertainty
        )

        obs, r, term, trunc, _ = env.step(a)
        positions.append(obs[0])
        rewards.append(r)

        if term or trunc:
            break

    return np.array(positions), np.array(rewards), (t + 1)
```

```python
# Episodio con modelo determinista (solo media del GP)
pos_det, rew_det, steps_det = run_episode(
    env, gps,
    use_variance=False,
    horizon=10,
    n_samples=20,
    goal_pos=0.45,
    seed=0
)

# Episodio con modelo que penaliza incertidumbre
pos_unc, rew_unc, steps_unc = run_episode(
    env, gps,
    use_variance=True,
    beta_uncertainty=1.0,  # prueba y ajusta
    horizon=10,
    n_samples=20,
    goal_pos=0.45,
    seed=0
)

print("Determinista: pasos =", steps_det, "   Cumulative reward =", rew_det.sum())
print("Con incertidumbre: pasos =", steps_unc, "   Cumulative reward =", rew_unc.sum())
```

In the deterministic case (using only the GP mean), the controller assumes the model is perfect across the entire state space. This can lead to planning trajectories that traverse regions where the GP has very little data, and therefore unreliable predictions. Although it sometimes still solves the task, it may require more steps or exhibit less stable behavior, especially when the model is not yet well-trained.

By incorporating the GP variance into the cost function, we explicitly penalize trajectories that pass through areas of high model uncertainty. This pushes the controller to prefer trajectories that remain within regions where the GP has seen more data and is more accurate. In practice, this can translate into more robust control, with more consistent trajectories and, in some cases, fewer steps to reach the summit, because the planner no longer "blindly trusts" very noisy predictions.

From a sampling efficiency perspective, using the GP uncertainty helps to better exploit the available information: the controller aggressively uses the model only where it has confidence in it. Depending on how this penalty is combined with the collection of new data, it can also guide exploration toward useful but little-known regions, improving model quality with fewer interactions. However, if the penalty is excessive, the agent may become overly conservative and move very little, so adjusting the uncertainty weight is crucial.
